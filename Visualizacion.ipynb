#1.Matplotlib

Instalar las biliotecas:

```python
import matplotlib.pyplot as plt
import numpy as np
```

##1.1. Gráfico de líneas

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 8, 7]

plt.plot(x, y, color='blue', linestyle='--', marker='o')
plt.title('Gráfico de Línea')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

##1.2. Gráfico de barras

Comparar cantidades entre categorías

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 17, 35, 29]

plt.bar(categorias, valores, color='skyblue')
plt.title('Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad')
plt.show()
```

##1.3. Histograma

Ver la distribución de datos numéricos.

```python
datos = np.random.normal(50, 10, 100)

plt.hist(datos, bins=10, color='orange', edgecolor='black')
plt.title('Distribución de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

##1.4. Gráfico de dispersión (scatter plot)

Analizar la relación entre dos variables.

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='red')
plt.title('Relación entre Variables')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```

##1.5. Boxplot (diagrama de caja)

Identificar valores atípicos y distribución.

```python
data = [7, 15, 13, 18, 22, 29, 35, 50, 100]

plt.boxplot(data)
plt.title('Boxplot de Ingresos')
plt.ylabel('Ingreso ($)')
plt.grid(True)
plt.show()
```

##1.6. Múltiples gráficos en una figura

Comparar visualmente varias variables o grupos.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

#2.Seaborn

Importar las bibliotecas:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Puede cargarse un conjunto de datos de ejemplo:

```python
df = sns.load_dataset('tips')
df.head()
```

##2.1. Gráfico de dispersión (scatterplot)

Relación entre dos variables cuantitativas.

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
plt.title('Cuenta Total vs Propina')
plt.show()
```

##2.2. Gráfico de regresión (regplot)

Muestra una regresión lineal ajustada.

```python
sns.regplot(data=df, x='total_bill', y='tip')
plt.title('Regresión Lineal - Cuenta vs Propina')
plt.show()
```

##2.3. Boxplot

Distribución y valores atípicos por categoría.

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='day', palette='Set2', legend=False)
plt.title('Distribución de Cuenta Total por Día')
plt.show()
```

##2.4. Swarmplot

Distribución de puntos sin superposición.

```python
sns.swarmplot(data=df, x='day', y='tip', hue='sex')
plt.title('Propinas por Día y Sexo')
plt.show()
```

##2.5. Histograma y KDE

Visualización de la distribución de una variable.

```python
sns.histplot(data=df, x='total_bill', kde=True, color='purple')
plt.title('Distribución de Cuenta Total')
plt.show()
```

##2.6. Mapa de calor (Heatmap)

Matriz de correlación.

```python
correlacion = df.corr(numeric_only=True)
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

##2.7. Pairplot

Relaciones entre todas las variables numéricas.


```python
sns.pairplot(df, hue='sex')
plt.suptitle('Relaciones entre Variables', y=1.02)
plt.show()
```

## 8. Gráfico categórico generalizado (Catplot)

Permite combinar diferentes tipos de gráficos por categoría.

```python
sns.catplot(data=df, x='day', y='total_bill', kind='violin', hue='sex', split=True)
plt.title('Distribución de Cuenta Total por Día y Sexo')
plt.show()
```

#3.Plotly

Importar las bilbiotecas:

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import plotly.figure_factory as ff
```

Emplearesmos  uno de los conjuntos de datos **Iris** incorporado en **Plotly**. El conjunto de datos **Iris** contiene información sobre la longitud y el ancho de los sépalos y pétalos de varias especies de iris.

Cargar el Conjunto de Datos Iris que contiene las siguientes columnas:
- `sepal_length`: Longitud del sépalo
- `sepal_width`: Ancho del sépalo
- `petal_length`: Longitud del pétalo
- `petal_width`: Ancho del pétalo
- `species`: Especie de la flor (Setosa, Versicolor, Virginica)


```python
df = px.data.iris()
df.head()
```

##3.1. Gráfico de Dispersión (Scatter Plot)

Gráfico de dispersión que muestre la relación entre el `sepal_length` y el `sepal_width` para cada especie de iris.

```python
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species", title="Gráfico de Dispersión de Iris")
fig.show()
```

##3.2. Gráfico de Barras

Gráfico de barras que muestre cuántas muestras hay para cada especie de iris.

```python
fig = px.bar(df, x="species", title="Distribución de Especies en el Conjunto Iris")
fig.show()
```

##3.3. Gráfico de Tortas (Pie Chart)

Gráfico para visualizar las proporciones de cada especie.

```python
fig = px.pie(df, names="species", title="Distribución de Especies (Pie Chart)")
fig.show()
```

##3.4. Histograma

Gráfico para ver la longitud del sépalo (`sepal_length`).

```python
fig = px.histogram(df, x="sepal_length", color="species", title="Histograma de Longitud del Sépalo")
fig.show()
```

##3.5. Mapa de Calor de la Matriz de Correlación

Gráfico de mapa de calor para mostrar la matriz de correlación entre las diferentes variables numéricas del conjunto de datos.


```python
# Calcular la matriz de correlación solo para columnas numéricas
corr_matrix = df.corr(numeric_only=True)

# Crear el mapa de calor
fig = ff.create_annotated_heatmap(z=corr_matrix.values,
                                  x=list(corr_matrix.columns),
                                  y=list(corr_matrix.columns),
                                  colorscale="Viridis")
fig.update_layout(title="Matriz de Correlación - Conjunto Iris")
fig.show()
```

##3.6. Combinar Gráficos - subgráficos

Podemos combinar múltiples gráficos en una sola figura usando subgráficos. Vamos a crear un gráfico con dos subgráficos: uno de barras para la distribución de especies y otro de dispersión para las longitudes del sépalo y el pétalo.



```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Crear subgráfico con dos subgráficos (filas=1, columnas=2)
fig = make_subplots(rows=1, cols=2, subplot_titles=["Distribución de Especies", "Dispersión Sepal vs Petal"])

# Gráfico de barras
bar_fig = go.Bar(x=df['species'].value_counts().index, y=df['species'].value_counts().values)
fig.add_trace(bar_fig, row=1, col=1)

# Gráfico de dispersión
scatter_fig = go.Scatter(x=df['sepal_length'], y=df['petal_length'], mode='markers', marker=dict(color=df['species'].apply(lambda x: 1 if x == "setosa" else (2 if x == "versicolor" else 3))))
fig.add_trace(scatter_fig, row=1, col=2)

# Actualizar diseño
fig.update_layout(title_text="Subgráfico de Distribución de Especies y Dispersión Sepal vs Petal", height=500, width=1000)
fig.show()
```
