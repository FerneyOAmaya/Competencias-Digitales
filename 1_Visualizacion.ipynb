## 1.Tipos de variables.

Las variables pueden representar datos **cuantitativos** o **cualitativos**. 

Las variables cuantitativas a su vez pueden representar datos **discretos** o **continuos**.

Las variables cuanlitativas o categóricas representan datos discretos que pueden ser ordenados (ordinal) o no ordenados (nominal).

Para seleccionar la mejor forma de graficar datos es importante identificar:
- ¿discretos o continuos?
- ¿ordenados o no ordenados?

En la Tabla 1 se presenta la clasificación y algunos ejemplos:

|:-|:-|
|Numéricas discretas|Número de personas en una habitación: 8, 10, 15|
|Numéricas continuas|El valor de la temperatura ambiental (°C): 37.6, 40.8|
|Categóricas no ordenadas|Género: Masculino, Femenino|

## 2.Elementos gráficos para codificar información.

La variación de los datos de una variable puede codificarse en un gráfico de la siguiente forma:
- por posición.
- por forma.
- por tamaño.
- por color.
- por las características de la línea.

Aunque existen múltiples sistemas de coordenadas, en este documento se abordará el sistema cartesiano en dos dimensiones (2D), donde $x$ representa la variable del eje horizontal y $y$ la del eje vertical.

A manera de ejemplo, se grafican en un plano cartesiano los valores que aparecen en la Tabla 4.

|x (km)|y (km)|Texto|
|:-:|:-:|:-|
|1|7|Temperatura: $37°C$|
|2|2|Humedad: $30$%|
|5|8|PM 2.5: $14 \mu g/{m}^3$|
|6|3|Altura: $1200 m$|

<center>Tabla 4. Valores a llevar a un plano cartesiano.</center> 

A qué escenario pueden corresponder los datos de la Tabla 4?

En la Figura 1 se presentan los valores en el plano cartesiano, mostrando diferentes posibilidades de forma, color y tamaño.

```python
#@title  { display-mode: "form" }
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 5, 6])
y = np.array([7, 2, 8, 3])
siz = [50, 200, 500, 800]
mar = ['s','>','d','o']
col = ['#32a852','#c7bcd4','#cfca72','#cc4bac']
xpos = [0.8, 1.0, 1.5, 1.5]
tex = ['Temperatura: 37°C','Humedad: 30 %','PM 2.5: $14\mu g/m^{3}$','Altura:  1200m']

f, ax = plt.subplots(1, 2, figsize=(10, 6))
ax[0].axis([0, 8, 0, 10])
ax[0].set_xlabel('Etiqueta Eje x', fontsize=14)
ax[0].set_ylabel('Etiqueta Eje y', fontsize=14)
ax[0].set_title('Título de la Figura.', fontsize=14)
ax[0].grid(color='b', ls = '-.', lw = 0.25)
ax[0].xaxis.set_major_locator(plt.MaxNLocator(8))
ax[0].yaxis.set_major_locator(plt.MaxNLocator(10))
ax[0].xaxis.set_minor_locator(plt.MaxNLocator(16))
ax[0].yaxis.set_minor_locator(plt.MaxNLocator(20))
bbox_props = dict(boxstyle="larrow,pad=0.3", fc="cyan", ec="b", lw=2)
ax[0].text(0.35, 4, "Major tick", ha="left", va="center",
            size=12, bbox=bbox_props)
ax[0].text(0.35, 5.5, "Minor tick", ha="left", va="center",
            size=12, bbox=bbox_props)
ax[0].text(0.35, 1, "y mínimo", ha="left", va="center",
            size=12, bbox=bbox_props, rotation=45)
ax[0].text(0.35, 8.9, "y máximo", ha="left", va="center",
            size=12, bbox=bbox_props,rotation=-45)
bbox_props = dict(boxstyle="round,pad=0.3", fc="cyan", ec="b", lw=2)
ax[0].text(4, 5, "Rejilla", ha="left", va="center",
            size=12, bbox=bbox_props)

for i in range(len(x)):
  ax[1].scatter(x[i],y[i],s=siz[i],color=col[i], marker=mar[i])
  ax[1].annotate(tex[i], xy=(x[i], y[i]), xytext=(x[i]-xpos[i], y[i]+0.5),size=14)

ax[1].axis([0, 8, 0, 10])
ax[1].set_xlabel('Ubicación (km)', fontsize=14)
ax[1].set_ylabel('Ubicación (km)', fontsize=14)
#ax[1].set_title('Visualización datos Tabla 4.', fontsize=14)
ax[1].xaxis.set_major_locator(plt.MaxNLocator(8))
ax[1].yaxis.set_major_locator(plt.MaxNLocator(10))
ax[1].xaxis.set_minor_locator(plt.MaxNLocator(16))
ax[1].yaxis.set_minor_locator(plt.MaxNLocator(20))
ax[1].grid(color='b', ls = '-.', lw = 0.25)
f.suptitle('Figura 1. Elementos para la visuaización.',fontsize=14)
f.show()
```

## 3.Gráficas básicas

Las variables categóricas pueden visualizarse empleando uno de los siguientes tipos de figuras:
- **Diagrama de barras simples (bar):** para representar conteo o cantidad de etiquetas.
- **Diagrama de torta (pie)**: para representar proporciones o relaciones entre las categorías.

Las variables numéricas ya sean continuas o discretas pueden visualizarse empleando uno de los siguientes tipos de figuras:
- **Líneas (plot)**: representa variables continuas.
- **Dispersión (scatter)**: representa variables discretas o un subconjunto de una variable que es continua.

La distribución de las variables numéricas puede hacerse con un **histograma (hist)**.

Las gráficas de **cajas y bigotes (box)** son útiles para presentar características estadísticas de las variables.

Para visualizar la relación entre dos variables puede emplearse un gráfico de **dispersión**.

#### 3.1.1. Gráfico de barras

En la Figura 2 se presenta la información en un gráfico de barras vertical.

```python
#@title  { display-mode: "form" }
habitantes1 = np.array([4273722, 1392623, 42892]);
grupos1 = ['Afrocolombiano', 'Indígenas', 'Otros'];
habitantes2 = np.array([30565, 7470, 4857]);
grupos2 = ['Raizal', 'Palenquero', 'Rrom'];

cmap = plt.get_cmap('Accent')

f, axes = plt.subplots(2,1, dpi=80, figsize=(6, 10))
rect0 = axes[0].bar(grupos1, habitantes1/1e6, color=cmap(0))
axes[0].set_ylabel('Habitantes en millones', fontsize=12)
axes[0].set_title('Distribución por grupos étnicos', fontsize=12)
axes[0].set_xticklabels(grupos1 , fontsize=12)
axes[0].set_ylim([0,5])

for rect in rect0:
 height = rect.get_height()
 axes[0].text(rect.get_x() + rect.get_width()/2, 1.01*height,
                '%5.3f' % float(height), ha='center', va='bottom')

rect1 = axes[1].bar(grupos2, habitantes2/1e3,color=cmap(1))
axes[1].set_ylabel('Habitantes en miles', fontsize=12)
axes[1].set_title('Distribución otros grupos étnicos', fontsize=12)
axes[1].set_xticklabels(grupos2, fontsize=12)
axes[1].set_ylim([0,35])

for rect in rect1:
 height = rect.get_height()
 axes[1].text(rect.get_x() + rect.get_width()/2, 1.01*height,
                '%5.3f' % float(height), ha='center', va='bottom')
f.suptitle('Figura 2. Gráfico de barras con distribución de grupos étnicos en Colombia, Censo 2005.', fontsize=12)
f.show()
```

#### 3.1.2. Gráfico de torta

En la Figura 3 se presenta la información en un diagrama de torta.

```python
#@title  { display-mode: "form" }
def func(pct, allvals):
    absolute = float(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:.3f})".format(pct, absolute)
cmap = plt.get_cmap('Accent')
color = [cmap(i) for i in np.linspace(0, 1, 8)]
leg = ['Afrocolombiano','Indígenas','Otros']

f, axes = plt.subplots(1,2, dpi=80, figsize=(10, 6))
axes[0].pie(habitantes1/1e6 ,colors=color, autopct=lambda pct: func(pct, habitantes1/1e6),textprops=dict(color="w",size=13))
axes[0].set_title('Distribución por grupos étnicos\n(millones)', fontsize=12)
axes[0].legend(['Afrocolombiano','Indígenas','Otros'])

cmap = plt.get_cmap('Set2')  #Accent
color = [cmap(i) for i in np.linspace(0, 1, 8)]
axes[1].pie(habitantes2/1e3, colors=color, autopct=lambda pct: func(pct, habitantes2/1e3),textprops=dict(color="w",size=13))
axes[1].set_title('Distribución otros grupos étnicos\n(miles)', fontsize=12)
axes[1].legend(['Raizal','Palenquero','Rrom'])
f.suptitle('Figura 3. Gráfico de torta con distribución\n de grupos étnicos en Colombia, Censo 2005.', fontsize=12)
f.show()
```

#### 3.2.1. Gráfico de dispersión

En la Figura 4 se presenta un diagrama de dispersión para las variables de la Tabla 2.

```python
#@title  { display-mode: "form" }
T = np.array([23.92,23.76,25.15,22.79,23.96,25.24,23.25,22.75])
PM = np.array([23.694533,11.199788,12.408048,13.856680,16.877331,33.068680,19.709193,8.3128980])
Altura = np.array([1775,1584,1549,1725,1642,1538,1637,1506])

cmap = plt.get_cmap('Accent')
mar = ['.','v','^','s','>','d','o','<']

f, ax = plt.subplots(2, 1, figsize=(5, 9))
for i in range(len(T)):
  ax[0].scatter(T[i], PM[i], s=150, color=cmap(i), marker=mar[i],label=str(Altura[i]))

ax[0].legend(markerscale=0.8, title='Altura (m):')
ax[0].axis([0, 30, 0, 40])
ax[0].set_xlabel('Temperatura (°C)', fontsize=14)
ax[0].set_ylabel('PM 2.5 ($\mu g/m^{3}$)', fontsize=14)

Tp = np.mean(T)
arm = 100^(Altura - 1500)

ax[1].scatter(T-Tp, PM , (Altura - 1500)^2 ,color = cmap(1), alpha=0.7)

for i in range(len(T)):
 ax[1].annotate(Altura[i], xy=(T[i]-Tp, PM[i]), xytext=(T[i]-Tp-0.3, PM[i]+1.3),size=12)

ax[1].axis([-3, 3, 0, 40])
Texty = 'Temperatura (°C) respecto al promedio={:.1f}'.format(Tp)
ax[1].set_xlabel(Texty, fontsize=14)
ax[1].set_ylabel('PM 2.5 ($\mu g/m^{3}$)', fontsize=14)

f.suptitle('Figura 4. Visualización numérica: datos ambientales.', fontsize=14)
f.show()
```

Los diagramas de cajas y bigotes representan viasualmente las características estadísticas: mediana o cuartil 2 (Q2), primer cuartil (Q1) y el tercer cuartil (Q3). Estos valores se presentan en la Tabla 7 para las variables Temperatura y Humedad relativa de la Tabla 6.



```python
#@title  { display-mode: "form" }
import pandas as pd

T = np.array([25.5,23.0,21.6,20.6,21.2,22.9,22.3,22.0,22.2,22.7,23.3,23.5])
H = np.array([53.3,64.3,71.5,75.3,71.0,63.3,64.6,65.0,64.1,64.3,62.2,61.1])

df = pd.DataFrame({
    'T (°C)':[np.quantile(T,0.75), np.quantile(T,0.50), np.quantile(T,0.25)],
    'H (%)' :[np.quantile(H,0.75), np.quantile(H,0.50), np.quantile(H,0.25)]
},  index =['Q3','Q2','Q1'])
print('Tabla 7. Cuartiles para las variables\n Temperatura y Humedad.')
df
```

En la Figura 6 se presentan los diagramas de cajas y bigotes para las variables Temperatura y Humedad de la Tabla 6.

```python
f, axes = plt.subplots(1, 2, figsize=(10, 6))
axes[0].boxplot(T)
axes[0].set_ylabel('T (°C)', fontsize=14)
axes[1].boxplot(H)
axes[1].set_ylabel('H (%)', fontsize=14)
f.suptitle('Figura 6. Diagramas de cajas y bigotes para la Temperatura y Humedad,', fontsize=14)
f.show()
```
