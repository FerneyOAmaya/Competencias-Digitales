#Paso 3: Explora, visualiza y guarda

##1. Crea un Dataframe

Crea un Dataframe. Un DataFrame es la forma como desde Python se puede manipuar una Tabla empleando la biblioteca Pandas.

Para crear el Dataframe, copia y ejecuta el siguiente código:

```python
import pandas as pd
df = pd.read_excel(file_path)
```

##2. Explora el archivo

Para explorar el archivo desde el cuaderno, copia y ejecuta el siguiente código:

```python
df.describe()
```

##3. Visualiza

Para visualizar las ventas por año

```python
import matplotlib.pyplot as plt
# Convertir la columna de fecha a tipo datetime y extraer el año
df['Año'] = pd.to_datetime(df['Fecha']).dt.year
# Agrupar por año y sumar las ventas
ventas_por_año = df.groupby('Año')['Costo'].sum()
# Visualizar los datos
plt.figure(figsize=(10, 6))
ventas_por_año.plot(kind='bar')
plt.title('Ventas por Año')
plt.xlabel('Año')
plt.ylabel('Total de Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Guarda

Esta es la forma de guardar los cambios en el Drive:

```python
#Guarda los cambios en un nuevo archivo Excel en el Drive
df.to_excel('/content/drive/My Drive/Origen/DB_AnalisisDatos_modificado.xlsx', index=False)
```
