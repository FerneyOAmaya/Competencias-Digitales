## 1.Escenario: Temperatura corporal

Suponga que se mide la frecuencia cardíaca una vez al día a dos pacientes y se obtienen los siguientes valores:

- Paciente A: 72, 76, 74, 74, 72, 76 pulsaciones/minuto
- Paciente B: 68, 91, 59, 60, 90, 76 pulsaciones/minuto 

Se observará que aunque el promedio de los dos pacientes es igual (74 pulsaciones/minuto), la variabilidad no es igual, el Paciente A tiene medidas más estables, mientras el Paciente B tiene medidas que varían más ampliamente a través de los días.

## 2.Cálculo de la media

La media es una medida de tendencia central y tiene ecuaciones diferentes para una  **población** y para una **muestra**:

- Para una **población**, la media:
$$\mu = \frac{\sum X_{i}}{N}$$

- Para una **muestra**, la media:
$$\overline{x} = \frac{\sum x_{i}}{n}$$

donde:<br>
- $X_{i}$ son las observaciones de la **población**,
- $N$ es el número de observaciones de la **población**,
- $x_{i}$ son las observaciones de la **muestra**,
- $n$ es el número de observaciones de la **muestra**.

Al calcular el promedio de frecuencia cardíaca para los dos pacientes se obtiene un valor igual:
- Promedio Paciente A: $\mu_{A} = 74$ pulsaciones/minuto
- Promedio Paciente B: $\mu_{B} = 74$ pulsaciones/minuto

Sin embargo, como se indicó previamente, la variabilidad de la frecuencia cardíaca de los dos pacientes no es igual.

Se presenta mayor detalle en el cálculo del promedio de la frecuencia cardíaca del Paciente A:

| Número de <br>observación <br> $_{i}$ | Frecuencia<br>  cardíaca <br> $X_{i}$ |
|:-:|:-:|
| 1 | 72 |
| 2 | 76 |
| 3 | 74 |
| 4 | 74 |
| 5 | 72 |
| 6 | 76 |
|----|----|
|$\sum$ | 444|

Con $N=6$ el promedio es:
$$\mu = \frac{444}{6} = 74$$

## 3.Medidas de variabilidad

Las medidas de variabilidad son:
- El rango
- La varianza
- La desviación estándar

El **rango** se calcula con la siguiente ecuación:
$$rango = max(X_{i}) - min(X_{i})$$

Al calcular el rango de frecuencia cardíaca para los dos pacientes:
- Rango Paciente A: $rango_{A} =  4\,$ pulsaciones/minuto
- Rango Paciente B: $rango_{B} = 32\,$ pulsaciones/minuto

La varianza y la desviación estándar son medidas más útiles de variabilidad e indican que tan alejadas del valor promedio se encuentran las observaciones. La varianza se calcula diferente para una **población** y para una **muestra**:

- Para una **población**, la **varianza**:
$$\sigma^{2} = \frac{\sum (X_{i} - \mu)^{2}}{N}$$

- Para una **muestra**, la **varianza**:
$$s^{2} = \frac{\sum (x_{i} - \overline{x})^{2}}{n}$$

donde:<br>
- $X_{i}$ son las observaciones de la **población**,
- $N$ es el número de observaciones de la **población**,
- $x_{i}$ son las observaciones de la **muestra**,
- $n$ es el número de observaciones de la **muestra**.

La **desviación estándar** para la población es $\sigma$ y para una muestra es $s$.

La varianza de la frecuencia cardíaca para los pacientes:
- Varianza Paciente A: $\sigma^{2} = 2.7\,$ (pulsaciones/minuto)$^{2}$
- Varianza Paciente B: $\sigma^{2} = 167.7\,$ (pulsaciones/minuto)$^{2}$

La desviación estándar de la frecuencia cardíaca para los pacientes:
- Desviación estándar Paciente A: $\sigma = 1.6\,$ pulsaciones/minuto
- Desviación estándar B: $\sigma = 12.9\,$ pulsaciones/minuto

Se presenta mayor detalle en el cálculo de la varianza de la frecuencia cardíaca del Paciente A:

| $_{i}$  |  $X_{i}$  | $(X_{i}-\mu)^{2}$ |
|:-------:|:---------:|:-----------------:|
| 1       | 72        | 4                 |
| 2       | 76        | 4                 |
| 3       | 74        | 0                 |
| 4       | 74        | 0                 |
| 5       | 72        | 4                 |
| 6       | 76        | 4                 |
|---------|-----------|-------------------|
|$\sum$   | 444       | 16 |

Con $N = 6$, la varianza de la frecuencia cardíaca del Paciente A:
$$\sigma^{2} = \frac{16}{6} = 2.7$$

La desviación estándar:
$$s = \sqrt{\sigma^{2}} = \sqrt{2.7}= 1.6$$

¿Porqué se emplearon las ecuaciones para la **población**?

## 4.Cálculos empleando numpy de Python.

Puede emplearse la biblioteca **numpy** para realizar los cálculos:


```
import numpy as np
```

Se definen los arreglos con las observaciones:

```
A = np.array([72, 76, 74, 74, 72, 76]);
B = np.array([72, 91, 59, 60, 90, 72]);
```

Si las observaciones corresponden con la **Población**, las características medibles para el Paciente A son:

```
print('Media               : %4.2f' % np.mean(A));
print('Rango               : %4.2f' % (np.max(A) - np.min(A)));
print('Varianza            : %4.2f' % np.var(A));
print('Desviación estándar : %4.2f' % np.std(A));
```

Si las observaciones corresponden a una **Muestra**, las características medibles son:

```
print('Media               : %4.2f' % np.mean(A));
print('Rango               : %4.2f' % (np.max(A) - np.min(A)));
print('Varianza            : %4.2f' % np.var(A, ddof=1));
print('Desviación estándar : %4.2f' % np.std(A, ddof=1));
```

## 5.Histograma empleando pyplot

Para las observaciones de frecuencia cardíaca:
- Paciente A: 72, 76, 74, 74, 72, 76 pulsaciones/minuto
- Paciente B: 68, 91, 59, 60, 90, 76 pulsaciones/minuto

Se grafica el histograma:

```


import matplotlib.pyplot as plt
f, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(A,bins=100)
axes[0].set_title('Histograma Paciente A')
axes[0].set_ylabel('Frecuencia')
axes[1].hist(B,bins=100)
axes[1].set_title('Histograma Paciente B')
f.show()

```
